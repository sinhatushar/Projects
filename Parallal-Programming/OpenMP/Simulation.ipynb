```python
import numpy as np 
import pptk 
import time
```

```python
with open('file.txt','r') as f_:
    po = f_.readlines()

poi = []
for i, l in enumerate(po):
    poi.append(np.array(l.strip().split("\t")))
poi = np.array(poi)
```

```python
p = pptk.viewer(poi)
p.set(point_size=0.005)
```

```python
def viz(points):
    points = np.array(points)
#     p = pptk.viewer(points)
    p.load(points)
    p.set(point_size=0.005)
```

```python
with open('trajectory.txt','r') as fp:
    points = []
    for i,lines in enumerate(fp):
        if lines is "\n":
            viz(points)
            time.sleep(5)
            points = []
#             continue
        else:
            point = lines.strip().split(" ")
            points.append(np.array(point))
```
